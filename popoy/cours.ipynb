## Conditions
Les conditions servent a modifier l'execution d'un programme

### if 
La condition : si.
Elle lance le bout de code en dessous d'elle si la condition est respecté. Par exemple : 

```python
x = 4
y = 0

if x > 3:
    y = 1
```

Ici, $x > 3$ est vrai. Ainsi, $y=1$

### else
La condition : else. Elle se lance si la condition if d'avant n'a pas été vrai

```python
x = 4
y = 0

if x < 3:
    y = 1
else:
    y = 2
```

Ici $x < 2$ est faux. Ainsi $y=2$
